```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m=c=0
    
    Learning_rate=0.001
    iteration=120
    n=len(x)
    for i in range (iteration):
        y_pred=m*x+c
        
        cost=(1/n)*sum([value**2 for value in (y-y_pred)])
        pdm=-(2/n)*sum(x*(y-y_pred))
        cdm=-(2/n)*sum(y-y_pred)
        
        m=m-Learning_rate *pdm
        c=c-Learning_rate*cdm
        
        print("m value :",m)
        print("n value :",n)
        print("cost value :",cost)
        print("\n")
    
    
X=np.array([1,3])
Y=np.array([2,4])

gradient_descent(X,Y)
```
